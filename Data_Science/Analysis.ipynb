**actual preprocessing for this dataset
- drop the column 

```python
df.head()
```

```python
cols_to_drop= ['Type','Coverage','AREA','REG','DEV']  # for removing columns
df.drop(columns=cols_to_drop, inplace=True)     #inplace = true for updating the list
```

```python
df.rename({'OdName':'Country',
        'AreaName':'Continent',
        'RegName':'Region',
        'DevName':'Status',}, axis=1, inplace=True)
```

```python
df.head(1)
```

```python
df.columns = df.columns.astype(str)  #convert column name to string (specific to this data)
```

```python
year = list(map(str, range(1980,2014)))  #list of the year
print(year)
```

```python
df['Total'] = df[year].sum(axis=1)
df.head(3)
```

```python
df.set_index('Country', inplace=True)
```

```python
df.head(3)
```

```python
# filter data by continent
filter_europe = df['Continent'] == 'Europe'
eu_df = df[filter_europe].copy()
eu_df
```

```python
eu_df.sort_values(by='Total', ascending= False, inplace=True)
eu_df.head()[year].T.plot(kind='line', 
figsize=(20,5),
linewidth=2,
marker='o',
markersize=5, 
title='Top Countries with Immigration in Europe',
subplots='True',layout=(2,3))
```

```python
eu_df.sort_values(by='Total', ascending= True, inplace=True)
eu_df.head()[year].T.plot(kind='line', 
figsize=(20,5),
linewidth=3,
marker='o',
markersize=7, 
title='Top Countries with Immigration in Europe',subplots='True',layout=(3,2))
```

```python
eu_df.sort_values(by='Total', ascending= False, inplace=True)
eu_df.head(6)[year].T.plot(kind='line', 
figsize=(20,5),
linewidth=3,
marker='o',
markersize=7, 
title='Top Countries with Immigration in Europe',subplots='True',layout=(3,2))

plt.show()
```
