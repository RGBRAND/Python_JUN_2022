```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = np.random.randint(0,100,50)
print(data)
```

```python
plt.figure(figsize=(20,5))
plt.plot(data)
```

```python
plt.figure(figsize=(20,5))
plt.plot(data,
color='Red',
linestyle=':',
marker='s',
markersize=10)
```

```python
plt.plot?       # information
```

```python
plt.figure(figsize=(25,5), facecolor='yellow')
plt.plot(data, 'g--x')
plt.title("simple line plot data")
plt.xlabel("X axis Data")
plt.ylabel("Y Axis Data")
```

```python
plt.figure(figsize=(20,5))
plt.plot(data,
color='green',
linestyle='solid',
marker='s',
markersize=10)
plt.title("simple line plot data", fontsize=15,color='red')
plt.xlabel("X axis Data", fontsize=15,color='Orange')
plt.ylabel("Y Axis Data", fontsize=15, color='red')
```

```python
fig = plt.figure(figsize=(25,5))
plt.plot(data,'g--x')
name = ['A','B','C','D','E','F','Z']
plt.xticks(np.arange(1,50,7),labels=name,rotation=45,color='red',fontsize=16)
plt.show()

```

```python
data1 = np.random.randint(0,100,50)
data2 = np.random.randint(0,100,50)
fig = plt.figure(figsize=(25,5))
plt.plot(data1,label='data1')
plt.plot(data2,label='data2')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(25,5))
plt.plot(data,'g--x')
name = ['apple', 'banana','mango']
plt.xticks([10,25,48],labels=name,rotation=45,color='red',fontsize=16)
plt.show()
```

```python
data1 = np.random.randint(0,100,50)
data2 = np.random.randint(0,100,50)
fig = plt.figure(figsize=(25,5))
plt.plot(data1,'r:',label='data1')
plt.plot(data2,label='data2')
plt.legend(loc='lower right', fontsize=16)
plt.savefig("sample.png", bbox_inches='tight')
```

```python
plt.figure(figsize=(25,10))
plt.subplot(2,1,1)
plt.plot(data1,'r:', label='data1')
plt.subplot(2,1,2)
plt.plot(data2,label='data2')
plt.legend(loc='lower right', fontsize=16)
plt.show()

```

```python
fig, ax= plt.subplots(figsize=(25,10),nrows=2,ncols=1,sharex=True)
ax[0].plot(data1,'rx:', label='data1')
ax[1].plot(data2, 'bo--',label='data2')
ax[0].legend(loc='lower right',fontsize=16)
ax[1].legend(loc='lower right',fontsize=16)
ax[0].grid(True)
ax[1].grid(True)
plt.show()
```
